# Gaussian PDF and CDF


```python
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
norm.pdf(0)
```

```python
norm.pdf(0, loc=5 ,scale = 10)

```

```python
import numpy as np
```

```python
r = np.random.randn(10)
```

```python
norm.pdf(r)
```

```python
# log joint probability is advantageous than joint propability because the joint is computationally expensive
```

```python
norm.logpdf(r)
```

```python
norm.cdf(r)
```

```python
norm.logcdf(r)
```

# Gaussian Distribution

```python
r = np.random.randn(10000) 
```

```python
r= 10*np.random.randn(10000) + 5
```

```python
r = np.random.randn(10000 ,2)
plt.scatter(r[:,0],r[:,1])
plt.show()
```

```python
r[:,1] = 5*r[:,1]+2
```

```python
plt.scatter(r[:,0],r[:,1])
plt.show()
```

```python
plt.scatter(r[:,0],r[:,1])
plt.axis('equal')
plt.show()
```

```python
cov = np.array([[1,0.8],[0.8,3]])
from scipy.stats import multivariate_normal as mvn

```

```python
mu = np.array([0,2])
```

```python
r = mvn.rvs(mean=mu, cov = cov , size =1000)
```

```python
plt.scatter(r[:,0],r[:,1])
plt.axis('equal')
plt.show()
```

```python
r = np.random.multivariate_normal(mean=mu, cov = cov , size = 1000)
plt.scatter(r[:,0],r[:,1])
plt.axis('equal')
plt.show()
    
```
