# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## line chart

```python
x=np.linspace(0,10,10)
```

```python
y =np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("time")
plt.ylabel("some function of time")
plt.title("Chart")
plt.show()
```

```python
#same example with more points
x= np.linspace(0,10,100) 
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

# Scatter plot

```python
 import pandas as pd
```

```python
a = pd.read_csv('d.csv', header= None).values
```

```python
x = a[:,0]
y = a[:,1]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x_line = np.linspace(0,100,100)
```

```python
y_line = 2*x_line + 1
```

```python
plt.scatter(x,y)
plt.plot(x_line,y_line)
plt.show()
```

# Histogram

```python
plt.hist(x) 
plt.show()
```

```python
r = np.random.random(10000)
plt.hist(r)
plt.show()
```

```python
plt.hist(r,bins = 20)
plt.show()
```

```python
y_actual = 2*x +1
residuals = y-y_actual
```
